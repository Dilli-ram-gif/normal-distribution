```python
# This is the function to check if the given number is an even or odd number. 
def is_odd(number):
    if type(number) == int:
        if number%2 != 0:
            return "The number is odd."
        else:
            return "The number is even."
    else:
        print("Pagal hai kya")
for i in range(1,11):
    y = is_odd(i)
    print(y)
        
        
```

```python
# Using Default Argument
def multiplication(a=1, b=2, c=3):
    return a*b*c

answer = multiplication(2,3,4)
print(answer)
```

```python
# Using Positional Arguments
def power_function(x,y):
    """
    This function gives the product of two numbers x and y. 
    """
    return x**y

ans = power_function(3,2)
print(ans)
```

```python
# Using keyword arguments
def addition_function(first, second, third):
    return first + second + third

addition_function(second = 3, first = 3, third =9)
```

```python
# Using args in a function
def add_function(*args):
    sum_result = 0
    for i in args:
        sum_result = sum_result + i
    
    print(args)  # Basically, it gives a tuple 
    return sum_result
    
add_function(2,2,22,2)   
```

```python
# **kwargs allows us to pass any number of keyword arguments
# Keyword arguments mean that they contain a key-value pair, like a Python dictionary
```

```python
def display(**kwargs):
    for (key,value) in kwargs.items():
        print(key, '-->',value)

display(India='Delhi',Nepal="Kathmandu",Pakistan="Islamabad",China="Beijing")
```

# Variable Scope

```python
def g(y):
    print(x)
    print(x+1)

x = 5
g(x)
print(x)
```

```python
def f(y):
    x = 1
    x +=1
    print(x)

x = 5
f(x)
print(x)
```

```python

def h(y):
    global x
    x +=  1
    
x = 5
h(x)
print(x)
```

```python
def f(x):
    x = x+1
    print('in f(x): x = ',x)
    return x

x = 3
z = f(x)
print('in main program scope: z = ',z)
print('in main program scope: x = ',x)
```
