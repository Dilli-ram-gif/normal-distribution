```python
sns.set_theme(style = "white")
a = np.random.normal(100,2,1000000)
sns.histplot(x=a)  # for continuous distribution, we should use histogram to represent data
plt.show()
```

```python
sns.set_theme(style = "white")
a = np.random.normal(100,2,1000000)
sns.histplot(x=a, kde=True)  # for continuous distribution, we should use histogram to represent data
plt.show()
```
