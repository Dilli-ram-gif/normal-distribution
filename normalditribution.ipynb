### Perfume bottles are filled with the average volume of 150 cc and the standard deviation of 2 cc. What percent of bottles will have a volume greater than 153 cc ?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
norm.sf(153,150,2)
```

#### Perfume bottles are filled with the average volume of 150 cc and the standard deviation of 2 cc. What percent of bottles will have a volume between 148 and 152 cc?

```python
norm.sf(148, 150, 2) - norm.sf(152, 150, 2)
```

```python
x_ax = np.linspace(144,156,1300)
y_ax = norm.pdf(x_ax,150,2)
sns.lineplot(x=x_ax, y= y_ax)
plt.show()
```

```python
x_ax = np.linspace(144,156,1300)
y_ax = norm.pdf(x_ax,150,2)
sns.lineplot(x=x_ax, y= y_ax)
plt.axvline(153,color="red")
plt.show()
```
